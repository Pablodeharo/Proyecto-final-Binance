```python
import websocket
import json
import sqlalchemy
from sqlalchemy import create_engine, Column, Integer, Float, String, DateTime
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import declarative_base
from sqlalchemy.orm import sessionmaker
import datetime
```

```python
# Funci√≥n de devoluci√≥n de llamada para procesar mensajes recibidos
def on_message(ws, message):
    print("Mensaje recibido:", message)

# Funci√≥n de devoluci√≥n de llamada para manejar errores
def on_error(ws, error):
    print("Error:", error)

# Funci√≥n de devoluci√≥n de llamada para manejar cierre de conexi√≥n
def on_close(ws):
    print("Conexi√≥n cerrada")

# Funci√≥n de devoluci√≥n de llamada para manejar apertura de conexi√≥n
def on_open(ws):
    print("Conexi√≥n establecida correctamente")

# URL del servidor WebSocket de Binance
websocket_url = "wss://stream.binance.com:9443/ws/!ticker@arr"

# Crear una instancia de WebSocket y establecer los manejadores de eventos
ws = websocket.WebSocketApp(websocket_url,
                            on_message=on_message,
                            on_error=on_error,
                            on_close=on_close)
ws.on_open = on_open

# Iniciar la conexi√≥n WebSocket
ws.run_forever()
```

```python
# Definir la base de datos y la tabla para Bitcoin
BaseBTC = declarative_base()

class BitcoinTransaction(BaseBTC):
    __tablename__ = 'bitcoin_transactions'
    id = Column(Integer, primary_key=True)
    timestamp = Column(DateTime)
    price = Column(Float)
    volume = Column(Float)
    open_price = Column(Float)
    close_price = Column(Float)

# Definir la base de datos y la tabla para Ethereum
BaseETH = declarative_base()

class EthereumTransaction(BaseETH):
    __tablename__ = 'ethereum_transactions'
    id = Column(Integer, primary_key=True)
    timestamp = Column(DateTime)
    price = Column(Float)
    volume = Column(Float)
    open_price = Column(Float)
    close_price = Column(Float)

# Definir la base de datos y la tabla para Ripple
BaseXRP = declarative_base()

class RippleTransaction(BaseXRP):
    __tablename__ = 'ripple_transactions'
    id = Column(Integer, primary_key=True)
    timestamp = Column(DateTime)
    price = Column(Float)
    volume = Column(Float)
    open_price = Column(Float)
    close_price = Column(Float)

# Definir la base de datos y la tabla para Litecoin
BaseLTC = declarative_base()

class LitecoinTransaction(BaseLTC):
    __tablename__ = 'litecoin_transactions'
    id = Column(Integer, primary_key=True)
    timestamp = Column(DateTime)
    price = Column(Float)
    volume = Column(Float)
    open_price = Column(Float)
    close_price = Column(Float)
```
